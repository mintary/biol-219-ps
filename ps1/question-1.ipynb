```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
D = 10 # we set our diffusion coefficient (Âµ^2 /s)
```

```python
def mean_square_displacement(t, D, v_d):
    return 2 * D * t + (v_d * t) ** 2
```

```python
time = np.logspace(0, 5, 100000) # we create time values from 1 to 1000
velocities = [0, 0.1, 0.2, 0.5] # we generate multiple velocity values
```

```python

plt.figure(figsize=(10,6))

for v in velocities:
    plt.loglog(time, mean_square_displacement(time, D, v), label=f"v_d = {v} Âµm/s")

plt.xlabel("Time (s)")
plt.ylabel("Mean Square Displacement (Âµm^2)")
plt.title("Mean Square Displacement vs. Time at different velocities")
plt.legend()
plt.grid(True, which="both", ls="--")
plt.show()

plt.savefig("msd_graph.png")
```

```python
v = 1000

# select two points on the line to calculate slope
id1 = 0
id2 = 100

t1 = time[id1]
t2 = time[id2]

log_msd1 = np.log(mean_square_displacement(t1, D, v))
log_msd2 = np.log(mean_square_displacement(t2, D, v))

log_t1 = np.log(t1)
log_t2 = np.log(t2)

slope = (log_msd2 - log_msd1) / (log_t2 - log_t1)

print(slope)
```
