```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

p_length_df = pd.read_csv("persistence_length.csv")

print(p_length_df)
```

```python
p_length_microtubule = p_length_df.loc[0:10, "Persistence length (μm)"]
print(p_length_microtubule)
```

```python
p_length_actin = p_length_df.loc[10:20, "Persistence length (μm)"]
print(p_length_actin)
```

```python
length_actin = p_length_df.loc[0:10, "Length of curve (μm)"]
print(length_actin)
```

```python
length_microtubule = p_length_df.loc[10:20, "Length of curve (μm)"]
print(length_microtubule)
```

```python

fig = plt.figure(figsize=(10, 6))


plt.hist(p_length_microtubule)
plt.title("Distribution of persistence lengths of 10 microtubules")
plt.locator_params(axis="both", integer=True, tight=True)
plt.xlabel("Persistence length (μm)")
plt.ylabel("Number of microtubules")

plt.savefig("p_length_microtubule")
```

```python
fig = plt.figure(figsize=(10, 6))

plt.hist(p_length_actin)
plt.title("Distribution of persistence lengths of 10 actin filaments")
plt.locator_params(axis="both", integer=True, tight=True)
plt.xlabel("Persistence length (μm)")
plt.ylabel("Number of microtubules")
plt.savefig("p_length_actin")
```

```python
# mean and standard deviation

# for microtubules
print("Microtubule")
print(f"Mean: {np.average(p_length_microtubule)}")
print(f"Standard deviation: {np.std(p_length_microtubule)}")

print("---")

# for actin
print("Actin")
print(f"Mean: {np.average(p_length_actin)}")
print(f"Standard deviation: {np.std(p_length_actin)}")

```

```python
fig = plt.figure()


data = [p_length_microtubule, p_length_actin]
ax = fig.add_axes([0, 0, 1, 1])

plt.boxplot(data)
plt.show()
```
