```python
import numpy as np
import matplotlib.pyplot as plt

V_avg = 1
f = 2

def g1(V_n):
    return V_n * (f - 1)

def g2(V_n):
    return 2 * f * V_avg + V_n * (1 - f)

V_n = np.linspace(0, 5, 1000)

g1_values = g1(V_n)
g2_values = g2(V_n)

plt.figure(figsize=(10, 6))
plt.plot(V_n, g1_values, label=r"$g_+(V_n) = V_n (f - 1)$", color='blue')
plt.plot(V_n, g2_values, label=r"$g_-(V_n) = 2f<V> + V_n (1 - f)$", color='orange')

plt.fill_between(V_n, g1_values, g2_values, where=(g1_values > g2_values), 
                 interpolate=True, color='skyblue', alpha=0.5, label="Area")
plt.fill_between(V_n, g1_values, g2_values, where=(g1_values < g2_values), 
                 interpolate=True, color='salmon', alpha=0.5, label="Area")

plt.title("$V_n$ vs. $g(V_n)$ in a linear growth model")
plt.xlabel("$V_n$")
plt.ylabel("$g(V_n)$")
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
def g1_log(V_n):
    return np.log(f)

def g2_log(V_n):
    return np.log(f* (2* V_avg - V_n) / V_n )

g1_log_values = np.full_like(V_n, np.log(f))
g2_log_values = g2_log(V_n)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(V_n, g1_log_values, label=r"$g_+(V_n) = \ln(f)$", color='blue')
plt.plot(V_n, g2_log_values, label=r"$g_-(V_n) = \ln(f(2<V> - V_n) / V_n)$", color='orange')

plt.fill_between(V_n, g1_log_values, g2_log_values, where=(g1_log_values > g2_log_values), 
                 interpolate=True, color='skyblue', alpha=0.5, label="Area")
plt.fill_between(V_n, g1_log_values, g2_log_values, where=(g1_log_values < g2_log_values), 
                 interpolate=True, color='salmon', alpha=0.5, label="Area")

plt.title("$V_n$ vs. $g(V_n)$ in an exponential growth model")
plt.xlabel("$V_n$")
plt.ylabel("$g(V_n)$")
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
